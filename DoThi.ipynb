```python
#a) Vẽ đồ thị của hàm y=sin(x) trên đoạn x∈[0,2π]. 
#b) Đặt tiêu đề cho biểu đồ là "Đồ thị hàm Cosine". 
#c) Trục x hiển thị từ 0 đến 2π, trục y hiển thị từ -1 đến 1. 
#d) Đặt nhãn trục x là "Góc (radian)" và trục y là "Giá trị Cosine"
import matplotlib.pyplot as plt
import numpy as np

# Giá trị x từ 0 đến 2π
x = np.linspace(0, 2 * np.pi, 1000)
# Giá trị y là sin(x)
y = np.sin(x)

# Vẽ đồ thị
plt.plot(x, y)
plt.title("Đồ thị hàm Cosine")
plt.xlabel("Góc (radian)")
plt.ylabel("Giá trị Cosine")

# Thiết lập phạm vi hiển thị cho trục x và y
plt.xlim(0, 2 * np.pi)
plt.ylim(-1, 1)

plt.show()
```

```python
#Vẽ đồ thị của hàm y= 1/x trên đoạn x∈[−10,−0.1]∪[0.1,10]
#Yêu cầu:
#Đặt tiêu đề cho biểu đồ là "Đồ thị hàm y= 1/x".
#Trục x hiển thị từ -10 đến 10 (ngoại trừ vùng [−0.1,0.1]), trục y hiển thị từ -10 đến 10.
#Đặt nhãn trục x là "Giá trị x" và trục y là "Giá trị y".
import matplotlib.pyplot as plt
import numpy as np

# Tạo các giá trị x
x1 = np.linspace(-10, -0.1, 500)
x2 = np.linspace(0.1, 10, 500)
# Tạo các giá trị y tương ứng với x
y1 = 1 / x1
y2 = 1 / x2

# Vẽ đồ thị
plt.plot(x1, y1, label='y = 1/x', color='blue')
plt.plot(x2, y2, label='y = 1/x', color='blue')
plt.title("Đồ thị hàm y= 1/x")
plt.xlabel("Giá trị x")
plt.ylabel("Giá trị y")

# Thiết lập phạm vi hiển thị cho trục x và y
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Đường lưới và chú thích
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```

```python
#Vẽ hình sin và cos trên cùng một biểu đồ 
#a) Vẽ đồ thị của hai hàm số y = sin(x) và y = cos(x) trên đoạn x in [0, 2π] . 
#b) Đặt tiêu đề biểu đồ là "Đồ thị của hàm sin và cos". 
#c) Đặt nhãn trục x là "Góc (radian)" và trục y là "Giá trị hàm số".
import matplotlib.pyplot as plt
import numpy as np

# Giá trị x từ 0 đến 2π
x = np.linspace(0, 2 * np.pi, 1000)
# Giá trị y cho sin(x) và cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Vẽ đồ thị
plt.plot(x, y_sin, label='y = sin(x)', color='blue')
plt.plot(x, y_cos, label='y = cos(x)', color='red')
plt.title("Đồ thị của hàm sin và cos")
plt.xlabel("Góc (radian)")
plt.ylabel("Giá trị hàm số")

# Đường lưới và chú thích
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```
